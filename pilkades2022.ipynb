```python
import csv
import pandas as pd
```

```python
pilkades_df = pd.read_csv('data/Pilkades2022.csv', sep=';')
pilkades_df.head(3)
```

```python
pilkades_df.rename(columns={'Kabupaten':'nama_kab', 'Kecamatan':'nama_kec','Desa':'nama_desa'}, inplace=True)
pilkades_df[['nama_kab', 'nama_kec', 'nama_desa']] = pilkades_df[['nama_kab', 'nama_kec', 'nama_desa']].astype(str).apply(lambda col: col.str.upper())
pilkades_df
```

```python
pilkades_df.info()
```

```python
desa_id = pd.read_csv('data/idwilayah.csv', dtype='str')
desa_id = desa_id[desa_id['nama_provinsi']=='JAWA TENGAH']
desa_id['nama_kab'] = desa_id['nama_kab'].str.replace('KAB. ', '')
desa_id.head(3)
```

```python
# desa_id.rename(columns={'nama_provinsi':'prov', 'nama_pemda':'kab', 'nama_desa':'desa'}, inplace=True)

desa_id
```

```python
df_merge = pilkades_df.merge(desa_id, how='left', on=['nama_kab', 'nama_kec', 'nama_desa'])
df_merge
```
